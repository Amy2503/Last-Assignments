```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import re

nltk.download('vader_lexicon')
```

```python
df = pd.read_csv('tweets-data.csv').head(500)
```

```python
def sentiment_scores(sentences):
  sid_obj=SentimentIntensityAnalyzer()
  sentiment_dict = sid_obj.polarity_scores(sentences)
  print("Sentence sentiment analysis: ", sentiment_dict)
  print(sentiment_dict['neg']*100, "%Negative")
  print(sentiment_dict['neu']*100, "%Neutrel")
  print(sentiment_dict['pos']*100, "%Positive")
  print("Sentence Overall Rated As: ", end = " ")

  compound = sentiment_dict['compound']
  if compound >= 0.05:
        label = "Positive"
  elif compound <= -0.05:
        label = "Negative"
  else:
        label = "Neutral"

  return pd.Series([label, compound])
```

```python
def clean_tweet(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.strip()
    return text

df['cleaned_tweet'] = df['Tweets'].astype(str).apply(clean_tweet)
df.head()
```

```python
df[['sentiment_label', 'sentiment_score']] = df['cleaned_tweet'].apply(sentiment_scores)

df.head()
```
