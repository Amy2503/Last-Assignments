```python
from transformers import pipeline
import pandas as pd
import re
```

```python
df = pd.read_csv('tweets-data.csv').head(500)
```

```python
def clean_tweet(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.strip()
    tokens = text.split()
    if len(tokens) > 250:
        tokens = tokens[:250]
    return ' '.join(tokens)

df['cleaned_tweet'] = df['Tweets'].astype(str).apply(clean_tweet)
```

```python
sentiment_classifier = pipeline('sentiment-analysis')
results = df['cleaned_tweet'].apply(sentiment_classifier)
df['sentiment_label'] = results.apply(lambda x: x[0]['label'])
df['sentiment_score'] = results.apply(lambda x: x[0]['score'])

df.head()
```
